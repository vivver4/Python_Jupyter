```python
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt

df=pd.read_csv('../RawData/1Qi_dianwei.csv', index_col=0)
df_pricereturn=df.pct_change().dropna()

std_list=[]
##这里46个数据i只取36，不然好像到后来窗口里面的数字会越来越少，可能导致波动率不准确
for i in range(36):
    std=df_pricereturn[i: i+10].std().values[0]
    std_list.append(std)
    
std_list=pd.Series(std_list, index=df.index[0:36])
std_list.plot(label='1期')
plt.legend()
```
