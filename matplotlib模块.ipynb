# 基本使用

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-1, 1, 50)
y1=2*x+1
y2=x**2
```

1. 设置曲线的颜色，宽度，类型

```python
plt.plot(x, y1)          #第一个是横坐标的值，第二个是纵坐标的值
plt.plot(x, y2, color='red', linewidth=1, linestyle='--')       
```

2. 设置label

```python
plt.plot(x, y1, label='linear line')        
plt.plot(x, y2, label='square line')  
plt.legend(loc='best')
```

# 坐标轴的调整

1. 设置坐标轴范围

```python
plt.xlim((-1,2))         
plt.ylim((-2,3))

plt.plot(x, y1)          
plt.plot(x, y2)     
plt.show()
```

2. 设置坐标轴名称

```python
plt.xlabel('I am x')    
plt.ylabel('I am y') 

plt.plot(x, y1)          
plt.plot(x, y2)     
plt.show()
```

3. 设置坐标轴刻度以及刻度名称 

```python
plt.xlim((-1, 2))
plt.ylim((-2, 3))
new_ticks=np.linspace(-1, 2, 5)
plt.xticks(new_ticks)                          
plt.yticks([-2, -1.8, -1, 1.22, 3], ['$really\bad$', '$bad$', '$normal$', '$good$', '$really\$good$'])

plt.plot(x, y1)          
plt.plot(x, y2)        
plt.show()
```

4. 使用plt.gca()获取当前坐标轴信息，使用.spines设置边框

```python
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))       #把左侧边框（也就是y轴进行移动）       

plt.plot(x, y1)          
plt.plot(x, y2)        
plt.show()
```

5. 使用axis.set_ticks_position设置坐标轴刻度或名称的位置：bottom, top, both等

```python
plt.xlim((-1, 2))
plt.ylim((-2, 3))
new_ticks=np.linspace(-1, 2, 5)
plt.xticks(new_ticks)                          
plt.yticks([-2, -1.8, -1, 1.22, 3], ['$really\bad$', '$bad$', '$normal$', '$good$', '$really\$good$'])

ax=plt.gca()
ax.xaxis.set_ticks_position('top')
ax.yaxis.set_ticks_position('right')


plt.plot(x, y1)          
plt.plot(x, y2)     
plt.show()
```

# Annotation标注
>__实例__说明

```python
#绘制直线并移动坐标
x=np.linspace(-3, 3, 50)
y=2*x+1
plt.plot(x, y)
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))  
ax.spines['bottom'].set_position(('data',0))  

#添加注释annotate
x0=1
y0=2*x0+1
plt.plot([x0, x0], [0, y0], 'k--')  #因为plt.plot()方法接受的x, y本来就是两个数组，这样传入的其实就是两个坐标(x0, 0), 以及(x0,y0)，这两个点的连线自然就垂直于x轴
plt.scatter([x0, ], [y0, ], s=50, color='r')    #同理上面，这里传入的其实就一个坐标点(x0, y0)，再进行设置，也就是目标点


plt.annotate(r'$2x+1=3$', xy=(x0, y0), xytext=(3,3.5), fontsize=18, arrowprops= \
             dict(arrowstyle='->', connectionstyle='arc3, rad=.5'))     #xy参数表示箭头顶端的坐标，xytext表示注释'2x+1=3'的起始坐标位置，arrowprops设置箭头

plt.text(-2.5, 3, r'$This\ is\ the\ some\ text.$', fontdict={'size':16, 'color':'r'})   #-2.5,3参数表示text起始的坐标
```
